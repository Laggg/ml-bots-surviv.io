```python
import numpy as np
import cv2
# conda install -c conda-forge opencv (in Anaconda prompt https://anaconda.org/conda-forge/opencv)
import pickle
import matplotlib.pyplot as plt                      
%matplotlib inline

def see_plot(pict, shape=(8,4)):
    #функция отрисовки 2D массива (всего кадра или его кропа/фрагмента)
    plt.figure(figsize=shape)
    plt.imshow(pict)
    plt.grid()
    plt.show()

def see_crop(p, cord):
    see_plot(p[cord[0]:cord[1],cord[2]:cord[3]])
    
def parse_video_to_shots(path_to_video):
    #функция считывания видео и разбиения его на кадры
    p = []
    cap = cv2.VideoCapture(path_to_video)
    i = 0
    while True:
        ret,frame=cap.read()
        if not ret:
        #if not ret or i == 500 #считать первые 500 кадров
            break
        gray_img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        p.append(gray_img)
        i+=1
    print("В этом видео найдено {} кадров".format(len(p)))
    return p
```

```python
%%time

# Допустим, вы скачали видео https://www.youtube.com/watch?v=ERDmPJ_0QJc&t, сохранили его на своем компьютере
# под именем videoplayback.mp4 в той же папке, где находится этот юпитер-ноутбук. Тогда для считывания кадров
# используем функцию parse_video_to_shots
picts = parse_video_to_shots("videoplayback.mp4")
#picts - список с чб кадрами. Кадр - 2D массив np.array c числами от 0 до 255  type=uint8
```
